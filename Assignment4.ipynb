### Seperate Dependent and Independent Variables

```python
# Now we will perform encoding
```

```python
x = dataset.iloc[:, dataset.columns != "Attrition"]
y = dataset.iloc[:,1:2]
```

### Encoding

```python
# Nominal One hot encoding
```

```python
# BusinessTravel column
BusinessTravel=pd.get_dummies(x["BusinessTravel"],drop_first=True)
```

```python
BusinessTravel
```

```python
x=pd.concat([x,BusinessTravel],axis=1)
```

```python
x.drop(["BusinessTravel"],axis=1,inplace=True)
```

```python
# Department column
Department=pd.get_dummies(x["Department"],drop_first=True)
Department
```

```python
x=pd.concat([x,Department],axis=1)
x.drop(["Department"],axis=1,inplace=True)
```

```python
x.head()
```

```python
# EducationField column
EducationField=pd.get_dummies(x["EducationField"],drop_first=True)
EducationField
```

```python
x=pd.concat([x,EducationField],axis=1)
x.drop(["EducationField"],axis=1,inplace=True)
```

```python
# Gender column
Gender=pd.get_dummies(x["Gender"],drop_first=True)
Gender
```

```python
x=pd.concat([x,Gender],axis=1)
x.drop(["Gender"],axis=1,inplace=True)
```

```python
# JobRole column
JobRole=pd.get_dummies(x["JobRole"],drop_first=True)
JobRole
```

```python
x=pd.concat([x,JobRole],axis=1)
x.drop(["JobRole"],axis=1,inplace=True)
```

```python
x.shape
```

```python
# MaritalStatus column
MaritalStatus=pd.get_dummies(x["MaritalStatus"],drop_first=True)
MaritalStatus
```

```python
x=pd.concat([x,MaritalStatus],axis=1)
x.drop(["MaritalStatus"],axis=1,inplace=True)
```

```python
x.shape
```

```python
# Over18 column
Over18=pd.get_dummies(x["Over18"],drop_first=True)
Over18
```

```python
x=pd.concat([x,MaritalStatus],axis=1)
x.drop(["Over18"],axis=1,inplace=True)
```

```python
#OverTime column
OverTime=pd.get_dummies(x["OverTime"],drop_first=True)
OverTime
```

```python
x=pd.concat([x,OverTime],axis=1)
x.drop(["OverTime"],axis=1,inplace=True)
```

```python
#Attrition column
Attrition=pd.get_dummies(y["Attrition"],drop_first=True)
Attrition
```

```python
y=pd.concat([y,Attrition],axis=1)
y.drop(["Attrition"],axis=1,inplace=True)
```

```python
y.head()
```

```python
#Ordinal Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
#Education column
x["Education"]=le.fit_transform(x["Education"])
```

```python
x["Education"]
```

```python
# EnvironmentSatisfaction column
x["EnvironmentSatisfaction"]=le.fit_transform(x["EnvironmentSatisfaction"])
```

```python
# JobInvolvement column
x["JobInvolvement"]=le.fit_transform(x["JobInvolvement"])
```

```python
# JobSatisfaction column
x["JobSatisfaction"]=le.fit_transform(x["JobSatisfaction"])
```

```python
# PerformanceRating column
x["PerformanceRating"]=le.fit_transform(x["PerformanceRating"])
```

```python
# RelationshipSatisfaction column
x["RelationshipSatisfaction"]=le.fit_transform(x["RelationshipSatisfaction"])
```

### Splitting Training and Testing DataSet

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Model Building using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
y_test
```

### Evaluation of Classification Model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

```python
probability=model.predict_proba(x_test)[:,1]
probability
```

```python
fpr,tpr,threshsholds = roc_curve(y_test,probability)
```

```python
plt.plot(fpr,tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={
 'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']

}
```

```python
grid_search=GridSearchCV(estimator=dtc,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
grid_search.best_params_
```

```python
dtc_cv=DecisionTreeClassifier(criterion= 'entropy',
 max_depth=3,
 max_features='sqrt',
 splitter='best')
dtc_cv.fit(x_train,y_train)
```
